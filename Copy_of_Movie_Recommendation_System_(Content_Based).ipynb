```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv (1).zip')
movies.head()
```

```python
# Step 3: Select useful features
movies = movies[['id','title','overview']]
movies.head()
```

```python
# Fill NaN with empty string
movies['overview'] = movies['overview'].fillna("NaN")
```

```python
# Step 4: Text Vectorization
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['overview']).toarray()
```

```python
# Step 5: Similarity Matrix
similarity = cosine_similarity(vectors)
```

```python
# Function to recommend movies
def recommend(movie):
    try:
        movie_index = movies[movies['title'] == movie].index[0]
        distances = similarity[movie_index]
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        recommended = [movies.iloc[i[0]].title for i in movie_list]
        return recommended   # ✅ direct list return karo
    except:
        return ["❌ Movie not found!"]
```

```python
print("🎬 Recommended Movies Like Avatar:")
recommend('Avatar')
```

```python
print("Recommended Movies Like The Dark Knight Rises")
recommend('The Dark Knight Rises')
```

```python
print("Recommended Movies Like Spectre")
recommend('Spectre')
```
